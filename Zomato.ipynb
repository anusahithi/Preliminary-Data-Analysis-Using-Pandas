```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Zomato.csv")
```

```python
data=data.drop(columns = ['url','address','reviews_list','menu_item','listed_in(type)','listed_in(city)','dish_liked'], axis=1)
```

```python
data.columns=['Name','Delivery', 'Table Booking','Rating', 'Votes','Phone_numbers', 'Location', 'Type','Cuisines','Cost for 2']
```

```python
data.shape
```

```python
data.isna().sum()
```

 ## 1. How many places are there in each location?

```python
ans_1=data.groupby('Location')['Name'].count().sort_values(ascending = False)
ans_1
```

## 2. How many restaurants/cafes have home delivery and Table Booking Option ?


```python
ans_2=data.groupby(['Table Booking','Delivery']).size().reset_index()
ans_2.rename(columns={0:'Count'},inplace=True)
```

```python
ans_2.iloc[3]
```

## 3. What are the restaurants that have a rating above 3.5 ?

```python
ans_3=data[data['Rating']>='3.5']
ans_3=ans_3[ans_3['Rating']!='NEW']
```

```python
ans_3.sort_values(by='Rating',ascending=False,inplace=True)
```

## 4. Which Location has the most number of good Restaurants ?

```python
ans_4=ans_3[['Name','Location','Rating']].reset_index()
```

```python
ans_4.groupby(['Location','Rating']).count().drop('Name', axis=1).sort_values(by='Rating', ascending=False).rename(columns={'index':'Count'})
```
